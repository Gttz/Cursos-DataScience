### Função Step (Degrau)

##### Usada apenas para problemas lineares separaveis (não usada em deep learning)
Apenas valores 0 ou 1, caso seja maior que 1 (exemplo 30) será 1, caso seja menor que 0 (-1) será 0

```python
def stepFunction(soma):
    if soma >=1:
        return 1
    return 0
```

```python
teste = stepFunction(30)
teste
```

```python
teste2 = stepFunction(-1)
teste2
```

### Função Sigmoid 

##### Usada para problemas de classificação binária
ela é mais usada para prevêr as probabilidades ou problemas de classificação (cachorro, gato)
sempre terá valores entre 0 e 1

```python
def sigmoidFunction(soma):
    return 1 / (1 + np.exp(-soma))
```

```python
teste = sigmoidFunction(30)
teste
```

```python
teste = sigmoidFunction(-1)
teste
```

```python
print(sigmoidFunction(0))
```

```python
print(sigmoidFunction(2.1))
```

### Hyperbolic tanget (função tangente hiperbólica)

##### Usada para classificação 
Sempre terá valores entre -1 e 1

```python
def tahnFunction(soma):
    return (np.exp(soma) - np.exp(-soma)) / (np.exp(soma) + np.exp(-soma))
```

```python
teste = tahnFunction(-.358)
teste
```

```python
print(tahnFunction(2.1))
```

### ReLU (rectified linear units)

#### Usadas em redes neurais convolucionais e profundas
Retorna apenas valores >=0

```python
def reluFunction(soma):
    if soma >= 0:
        return soma
    return 0
```

```python
teste = reluFunction(-0.358)
teste
```

```python
print(reluFunction(2.1))
```

### Softmax

##### Usada em problemas de classificação com três ou mais classes
Ela é usada quando se quer a previsão de 3 classes ou mais classes (verde, vermelho, azul) também retorna probababilidade das classes

```python
def softmaxFunction(x):
    ex = np.exp(x)
    return ex / ex.sum()
```

```python
# retorna a chance de cada classe.
valores = [5.0, 2.0, 1.3]
softmaxFunction(valores)
```
