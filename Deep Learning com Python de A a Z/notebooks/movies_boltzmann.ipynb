### Restricted Boltzmann Machines


```python
from rbm import RBM
import numpy as np
```

```python
rbm = RBM(num_visible = 6, num_hidden = 2)
```

###### Declaração dos filmes e treinamento do algorítimo

```python
df = np.array([[1,1,1,0,0,0], # Terror 3/3 e comédia 0/3
               [1,0,1,0,0,0], # Terror 2/3 e comédia 0/3
               [1,1,1,0,0,0], # Terror 3/3 e comédia 0/3
               [0,0,1,1,1,1], # Terror 1/3 e comédia 3/3
               [0,0,1,1,0,1], # Terror 1/3 e comédia 2/3
               [0,0,1,1,0,1]]) # Terror 1/3 e comédia 2/3

filmes = ["A bruxa", "Invocação do mal", "O chamado",
          "Se beber não case", "Gente grande", "American pie"]
```

```python
rbm.train(df, max_epochs = 5000)
```

```python
rbm.weights
```

Aqui podemos observar que o algoritmo conseguiu descobrir sozinho cada uma das categorias (terror e comédia), lembrando que não passamos quais eram as categorias dos filmes.

###### Criação dos usuários para os filmes de terror e comédia

```python
usuario1 = np.array([[1,1,0,1,0,0]])
usuario2 = np.array([[0,0,0,1,1,0]])
```

```python
# Foco em terror
rbm.run_visible(usuario1)
```

```python
# Foco em comédia
rbm.run_visible(usuario2)
```

Podemos observar que o algoritmo recomendou filmes de terror para o usuário 1 e filmes de comédia para o usuário 2

```python
# teste do usuário 1
camada_escondida1 = np.array([[0, 1]])
recomendacao1 = rbm.run_hidden(camada_escondida1)
recomendacao1
```

```python
# teste do usuário 2
camada_escondida2 = np.array([[1, 0]])
recomendacao2 = rbm.run_hidden(camada_escondida2)
recomendacao2
```

###### Filmes recomendados para cada usuário

```python
for i in range(len(usuario1[0])):
    # print(usuario1[0,i])
    if usuario1[0,i] == 0 and recomendacao1[0,i] == 1:
        print(filmes[i])
```

Com isso, observamos que para o usuário 1 foi recomendado o filme 'O chamado'

```python
for i in range(len(usuario2[0])):
    # print(usuario1[0,i])
    if usuario2[0,i] == 0 and recomendacao2[0,i] == 1:
        print(filmes[i])
```

Com isso, observamos que para o usuário 2 foi recomendado os filmes 'O chamado' e 'American pie'
