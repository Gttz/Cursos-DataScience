###### Separando os previsores e a classe

```python
# Definição das colunas que serão features
X = df.iloc[:, 0:6].values
# Definição da variável-alvo
y = df.iloc[:, 6].values
```

###### Transformação dos dados de saída de categórico para númerico

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

lbe = LabelEncoder()
```

```python
y = lbe.fit_transform(y)
y
```

###### Importações
Para isso, utilizaremos a biblioteca ``keras``

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.wrappers.scikit_learn import KerasClassifier
```

###### Divisão de treino e teste
75% para treinamento e 25% para teste

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y , test_size = 0.25)
```

###### Parâmetros da rede neural artificial:
Link: https://keras.io/

    Sequential - Criação de uma nova rede neural
    Dense - Todos os neurônios conectados
    units - Quantidade de neurônios que fazem parte da camada
    activation - Função de ativação que será inserida
    kernel_initializer - inicialização dos pesos
    input_dim - quantos elementos existem na camada de entrada
    Dropout - é usado para diminuir a chance de overfitting (20% dos neurônios de entrada são zerados)
   
Ponto de partida para o paramêtro ``units`` - número de entradas(6) +  número de saidas(1) / 2 = 3.5
    
###### Parâmetros do otimizador

    optimizer - descida do gradiente e descida do gradiente estocástica (adam é uma otimização do estocástica)
    loss - Função de perda (binary_crossentropy pois só há uma saída)
    metrics - Métrica de avaliação (pode ser colocado mais de um)

```python
classificador = Sequential()
# Camada de entrada
classificador.add(Dense(units = 60, activation = 'tanh', kernel_initializer = 'random_uniform', input_dim = 6))
classificador.add(Dropout(0.2))
# Camada oculta
classificador.add(Dense(units = 60, activation = 'tanh', kernel_initializer = 'random_uniform'))
classificador.add(Dropout(0.2))
# Camada oculta
classificador.add(Dense(units = 60, activation = 'tanh', kernel_initializer = 'random_uniform'))
classificador.add(Dropout(0.2))
# Camada de saida
classificador.add(Dense(units = 1, activation = 'sigmoid'))
# Ajustes finais
classificador.compile(optimizer = 'Adam', loss = 'binary_crossentropy', metrics = ['binary_accuracy'])
```

###### Parâmetros do treinamento

    batch_size - cálculo de erro a cada 10 registros
    epochs - quantidade de vezes que será feito o ajuste dos pesos

```python
classificador.fit(X_train, y_train, batch_size = 3, epochs= 500)
```

```python
resultado = classificador.evaluate(X_test, y_test)
resultado
```

```python
previsoes = classificador.predict(X_test)
previsoes = (previsoes > 0.5)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

precisao = accuracy_score(y_test, previsoes)
matriz = confusion_matrix(y_test, previsoes)
```
